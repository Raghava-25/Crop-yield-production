## Step 6: Outlier Detection
Identify and handle outliers using visualizations and statistical methods.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Box plot for outlier detection in Production üèÜüìä
plt.figure(figsize=(12, 6))
sns.boxplot(x=df['Temperature'])
plt.title('Box Plot for Temperature')
plt.show()

# Handling outliers (example: using IQR method for Production üèÜüìä)
q1_temp = df['Temperature'].quantile(0.25)
q3_temp = df['Temperature'].quantile(0.75)
iqr_temp = q3_temp - q1_temp

# Define outliers for Production üèÜüìä
outliers_temp = df[(df['Temperature'] < (q1_temp - 1.5 * iqr_temp)) | (df['Temperature'] > (q3_temp + 1.5 * iqr_temp))]
print("Outliers in Temperature:")
print(outliers_temp)
```

The box plot for temparature shows that the majority of the data points falls between 34 and 36 degrees with a few outliers below 26 degrees.

## Step 7: Hypothesis Testing
Formulate hypotheses based on initial observations and test them using statistical methods.

```python
from scipy.stats import ttest_ind

# Hypothesis: Crops in Maharashtra have different production compared to crops in Punjab
ap_crops = df[df['State_Name'] == 'Andhra Pradesh']['Production']
andaman_crops = df[df['State_Name'] == 'Andaman and Nicobar Islands']['Production']

# Perform t-test
t_stat, p_value = ttest_ind(ap_crops, andaman_crops)

print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

The t-test results show no significant difference in crop production between Andhra Pradesh and Andaman & Nicobar Islands.(p-value > 0.05).

## Step 8: Summarize Findings
Compile insights and key findings from the EDA into a report.
Use visualizations and statistical summaries to present your results.

```python
import matplotlib.pyplot as plt

# Group by 'Season üå±üå¶Ô∏è' (e.g., Kharif vs Rabi) and calculate mean for selected features
df.groupby('Season')[['Humidity', 'Temperature', 'Soil_Moisture']].mean().T.plot(kind='barh', figsize=(10, 10))

plt.title('Mean Values of Production, Temperature, and Soil Moisture by Season')
plt.xlabel('Mean Value')
plt.ylabel('Features')
plt.show()
```

**Key Insights on Crop Yield Prediction:**

**Seasonal Impact:**

Rabi and Kharif seasons have distinct effects on crop yield due to different weather patterns. For example, water-loving crops like rice do better in Kharif, while dry-season crops like wheat thrive in Rabi.

**Temperature:**

Crops have an optimal temperature range for growth. Too high or too low temperatures can reduce yield. Some crops may perform better in hotter conditions (e.g., maize), while others need cooler temperatures (e.g., wheat).

**Humidity:**

High humidity helps some crops (e.g., rice), but excessive moisture can promote diseases. Crops like wheat may prefer drier conditions.

**Soil Moisture:**

Optimal soil moisture is crucial for good crop growth. Both too much and too little moisture can harm yield. Irrigation may be necessary in areas with unpredictable rainfall.

**Area of Land:**

Larger land areas typically lead to higher production, but only if other factors like soil and climate are favorable. Efficient land use is key.

**Crop Type:**

Different crops have specific needs for water, temperature, and soil. For example, maize requires more water and heat than wheat.

**Yearly Trends:**

Analyzing production by year helps identify trends and anomalies, such as poor yields caused by weather events or diseases.

**Geographical Factors:**

Crops perform differently across states and districts due to variations in climate and soil. Tailored strategies are needed for different regions.

**Outliers:**

Extremely high or low yields might indicate unusual weather conditions or errors in data, and investigating these can help improve predictions.

**Correlations**

Factors like temperature, humidity, and soil moisture often show strong links to yield. Understanding these relationships helps identify what impacts yield most.

```python
data = df.drop(['State_Name','District_Name','Season','Crop'], axis = 1)
```

```python
data.corr()
```

```python
sns.heatmap(data.corr(), annot =True, fmt='.4f', cmap='coolwarm', vmax=1, vmin=-1)
plt.title('Correlation Matrix')
```

```python
dummy = pd.get_dummies(data)
dummy
```

<b><i> Splitting dataset into train and test dataset </i></b>

```python

from sklearn.model_selection import train_test_split

x = dummy.drop(["Production","Yield"], axis=1)
y = dummy["Production"]

# Splitting data set - 25% test dataset and 75% 

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25, random_state=42)

print("x_train :",x_train.shape)
print("x_test :",x_test.shape)
print("y_train :",y_train.shape)
print("y_test :",y_test.shape)
```

```python
from sklearn.metrics import mean_squared_error
lr_predict = model.predict(x_test)
scores_regr = mean_squared_error(y_test, lr_predict)
print(scores_regr)
# print(mean_squared_error(y_train, lr_predict))
```

```python
from sklearn.metrics import r2_score
r = r2_score(y_test,lr_predict)
print("R2 score : ",r)
```

```python
plt.scatter(x = np.arange(0,50), y = y_test[:50])
plt.scatter(x = np.arange(0,50), y = (lr_predict[:50]))
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Linear Regression')
```

Clearly, the dataset is not good for linear regression.

R2 score: This is pronounced as R-squared, and this score refers to the coefficient of determination. 
This tells us how well the unknown samples will be predicted by our model.

# <b> Decision Tree </b>

```python
# Training model 
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 5)
regressor.fit(x_train,y_train)

# Predicting results
decisiontree_predict = regressor.predict(x_test)
decisiontree_predict
```

```python
regressor.score(x_test,y_test)
```

```python
plt.scatter(x = np.arange(0,50), y = y_test[:50])
plt.scatter(x = np.arange(0,50), y = (decisiontree_predict[:50]))
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Decision Tree ')
```

```python
# Calculating R2 score :

from sklearn.metrics import r2_score
r2 = r2_score(y_test,decisiontree_predict)
print("R2 score : ",r2)
```

# **Random Forest Algorithm**

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators = 11)
model.fit(x_train,y_train)
rf_predict = model.predict(x_test)
rf_predict
```

```python
model.score(x_test,y_test)
```

```python
# Calculating R2 score

from sklearn.metrics import r2_score
r1 = r2_score(y_test,rf_predict)
print("R2 score : ",r1)
```

```python
# Calculating Adj. R2 score: 

Adjr2_1 = 1 - (1-r1)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
print("Adj. R-Squared : {}".format(Adjr2_1))
```

```python
ax = sns.distplot(y_test, hist = False, color = "r", label = "Actual value ")
sns.distplot(rf_predict, hist = False, color = "b", label = "Predicted Values", ax = ax)
plt.title('Random Forest Regression')
```

```python
plt.scatter(x = np.arange(0,50), y = y_test[:50])
plt.scatter(x = np.arange(0,50), y = (rf_predict[:50]))
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Random Forest')
```
